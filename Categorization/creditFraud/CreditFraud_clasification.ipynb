```python
import pandas as pd
import seaborn as sns
from sklearn.svm import SVC
import matplotlib.pyplot as plt
from sklearn.utils import shuffle

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import plot_confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, classification_report
```

```python
df = pd.read_csv("/content/drive/MyDrive/CieÌ‚ncias de dados rep/Categorization/creditFraud/creditcard.csv")
```

```python
df_0 = df[df["Class"] == 0]
df_1 = df[df["Class"] == 1]
```

```python
X = ['Time', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10',
     'V11', 'V12', 'V13', 'V14', 'V15', 'V16', 'V17', 'V18', 'V19', 'V20',
     'V21', 'V22', 'V23', 'V24', 'V25', 'V26', 'V27', 'V28', 'Amount']
y = ['Class']
test_d0 = df_0.sample(n=int(len(df_0)/3),random_state=42)
test_d1 = df_1.sample(n=int(len(df_1)/3),random_state=42)
test_df = shuffle(pd.concat([test_d0,test_d1])).reset_index(drop=True)
test_df_X = test_df[X]
test_df_y = test_df[y]

train_d0 = df_0[~df_0.index.isin(test_d0.index)]
train_d1 = df_1[~df_1.index.isin(test_d1.index)]
train_df = shuffle(pd.concat([train_d0,train_d1])).reset_index(drop=True)
train_df_X = train_df[X]
train_df_y = train_df[y]
```

```python
DTC = DecisionTreeClassifier(random_state=42)
KNN = KNeighborsClassifier()
LRC = LogisticRegression(random_state=42)
DTC.fit(train_df_X,train_df_y)
KNN.fit(train_df_X,train_df_y)
LRC.fit(train_df_X,train_df_y)
prev_dtc = DTC.predict(test_df_X)
prev_knn = KNN.predict(test_df_X)
prev_lrc = LRC.predict(test_df_X)
```

```python
print("DecisionTreeClassifier accuracy: {}".format(accuracy_score(test_df_y["Class"].values,prev_dtc)))
print("KNeighborsClassifier accuracy: {}".format(accuracy_score(test_df_y["Class"].values,prev_knn)))
print("LogisticRegression accuracy: {}".format(accuracy_score(test_df_y["Class"].values,prev_lrc)))
print("=="*50)
print("DecisionTreeClassifier confusion_matrix: \n{}".format(confusion_matrix(test_df_y["Class"].values,prev_dtc)))
print("KNeighborsClassifier confusion_matrix: \n{}".format(confusion_matrix(test_df_y["Class"].values,prev_knn)))
print("LogisticRegression confusion_matrix: \n{}".format(confusion_matrix(test_df_y["Class"].values,prev_lrc)))
print("=="*50)
print("DecisionTreeClassifier classification_report: \n{}".format(classification_report(test_df_y["Class"].values,prev_dtc)))
print("KNeighborsClassifier classification_report: \n{}".format(classification_report(test_df_y["Class"].values,prev_knn)))
print("LogisticRegression classification_report: \n{}".format(classification_report(test_df_y["Class"].values,prev_lrc)))
```
